```python
import random

# Definindo os compartimentos e as aranhas
compartimentos = [0, 1, 2, 3]
aranhas = [0, 3]

# Definindo as probabilidades de transição
probFicar = 0.4
probMover = 0.6
probMoverParaVizinho = probMover / 2

# Função para simular o movimento da mosca
def simularMovimento(inicial, passos):
    estado = inicial
    for _ in range(passos):
        if estado in aranhas:
            return estado
        movimento = random.random()
        if movimento < probFicar:
            continue
        elif movimento < probFicar + probMoverParaVizinho:
            estado = max(0, estado - 1)  # Mover para a esquerda, se possível
        else:
            estado = min(3, estado + 1)  # Mover para a direita, se possível
    return estado

# Função para calcular a probabilidade de ser capturada no terceiro minuto
def probabilidadeCapturaTerceiroMinuto(simulacoes):
    capturas = 0
    for _ in range(simulacoes):
        if simularMovimento(1, 3) in aranhas:
            capturas += 1
    return capturas / simulacoes

# Função para calcular o número médio de passos até a absorção
def numeroMedioPassosAbsorcao(simulacoes):
    totalPassos = 0
    for _ in range(simulacoes):
        estado = 1
        passos = 0
        while estado not in aranhas:
            movimento = random.random()
            if movimento < probFicar:
                pass
            elif movimento < probFicar + probMoverParaVizinho:
                estado = max(0, estado - 1)
            else:
                estado = min(3, estado + 1)
            passos += 1
        totalPassos += passos
    return totalPassos / simulacoes

# Função para calcular a probabilidade de absorção associada a cada estado
def probabilidadeAbsorcao(simulacoes, estadoInicial):
    absorcao = {0: 0, 3: 0}
    for _ in range(simulacoes):
        estado = estadoInicial
        while estado not in aranhas:
            movimento = random.random()
            if movimento < probFicar:
                pass
            elif movimento < probFicar + probMoverParaVizinho:
                estado = max(0, estado - 1)
            else:
                estado = min(3, estado + 1)
        absorcao[estado] += 1
    return {estado: absorcao[estado] / simulacoes for estado in absorcao}

# Número de simulações
simulacoes = 1000000

# Resultados
probCapturaTerceiroMinuto = probabilidadeCapturaTerceiroMinuto(simulacoes)
numeroMedioPassos = numeroMedioPassosAbsorcao(simulacoes)
probAbsorcaoInicial1 = probabilidadeAbsorcao(simulacoes, 1)
probAbsorcaoInicial2 = probabilidadeAbsorcao(simulacoes, 2)

print(f"Probabilidade de captura exatamente no terceiro minuto (inicial no compartimento 1): {probCapturaTerceiroMinuto:.4f}")
print(f"Número médio de passos para a absorção: {numeroMedioPassos:.4f}")
print(f"Probabilidade de absorção (inicial no compartimento 1): {probAbsorcaoInicial1}")
print(f"Probabilidade de absorção (inicial no compartimento 2): {probAbsorcaoInicial2}")
```
