# Reproduzindo resultados do artigo escolhido

#### Importando as bibliotecas

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

#### MÃ©todo para plotar as imagens

```python
def plot_image(image):
    chosen_curve = mpimg.imread(image)
    plt.imshow(chosen_curve)
    plt.axis('off')  
    plt.show()
```

#### Plotando a primeira curva escolhida

```python
plot_image("ChosenCurve1.png")
```

#### Reproduzindo a curva por meio de uma funÃ§Ã£o exponencial

```python
A = 96.31222146355297
B = -93.3607015459303
C = -0.023494455428541956

x = np.linspace(5, 800, 1000)
y =  A + B * np.exp(C * x)

plt.plot(x, y)
plt.ylim(0, 10**2)
plt.xlim(0, 800)
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('ReproduÃ§Ã£o por funÃ§Ã£o exponencial')
plt.show()
```

#### Plotando a segunda curva escolhida

```python
plot_image("ChosenCurve2.png")
```
