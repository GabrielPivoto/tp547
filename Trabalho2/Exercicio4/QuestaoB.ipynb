```python
### b) I = ∫ exp (x+x^2)dx
import numpy as np
import matplotlib.pyplot as plt

N = 100000

#gerar N amostras uniformes
u_Importancia = np.random.uniform(0, 1, N)
u_MonteCarlo = np.random.rand(N)

# Integral
def f(x):
    return np.exp(x + x**2)

# Intervalo de integração
a = -2
b = 2

# calcular a integral pelo método de Monte Carlo
x = a + (b - a) * u_MonteCarlo
integral_m = np.sum(f(x)) * (b - a) / N
print(f'Integral pelo método de Monte Carlo = {integral_m:.4f}')

# calcular a integral por Importância
x = a + (b - a) * u_Importancia
integral_i = np.sum(f(x)) * (b - a) / N
print(f'Integral pelo método de Importância = {integral_i:.4f}')


#plotar o grafico
x = np.linspace(a, b, N)
curve_y = f(x)
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, where=((x >= a) & (x <= b)), color='blue', alpha=0.3)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=a, color='black', linestyle='--', linewidth=0.5)
plt.axvline(x=b, color='black', linestyle='--', linewidth=0.5)
plt.xlim(a, b)
plt.xlabel('x')
plt.ylabel('y')
plt.text(0, 100, f'Integral (Monte Carlo) = {integral_m:.4f}\nIntegral (Importância) = {integral_i:.4f}', fontsize=12, color='red')

plt.show()
```
