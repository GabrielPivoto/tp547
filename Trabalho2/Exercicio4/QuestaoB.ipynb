```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000000

# Gerar N amostras uniformes
u_Importancia = np.random.uniform(0, 1, N)
u_MonteCarlo = np.random.rand(N)

# Calcular a integral pelo método de Monte Carlo
x = -np.log(1 - u_MonteCarlo)
integral_MonteCarlo = np.sum(np.exp((x + x**2)))/N
print(f'Integral pelo método de Monte Carlo = {integral_MonteCarlo:.4f}')

# Calcular a integral por Importância
x = -np.log(1 - u_Importancia)
integral_Importancia = np.sum(np.exp((x + x**2)))/N
print(f'Integral pelo método de Importância = {integral_Importancia:.4f}')

# Plotar o gráfico
x = np.linspace(-2, 2, N)
curve_y = np.exp(x + x**2)
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.xlim(-2, 2)
plt.xlabel('x')
plt.ylabel('y')
plt.text(4, 10000, f'Integral (Monte Carlo) = {integral_MonteCarlo:.4f}\nIntegral (Importância) = {integral_Importancia:.4f}', fontsize=12, color='red')

plt.show()
```
