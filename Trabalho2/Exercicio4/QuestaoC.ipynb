```python
### c) I = ∫ x(1 + x^2)^{-2}dx
import numpy as np
import matplotlib.pyplot as plt

N = 100000

#gerar N amostras uniformes
u_Importancia = np.random.uniform(0, 1, N)
u_MonteCarlo = np.random.rand(N)

x = -np.log(1 - u_MonteCarlo)
#calcular a integral pelo método de Monte Carlo
integral = np.sum(x*pow(1 + x**2 , -2)/np.exp(-x))/N
print(f'Integral pelo método de Monte Carlo = {integral:.4f}')

x = -np.log(1 - u_Importancia)
#calcular a integral por Importância
integral = np.sum(x*pow(1 + x**2 , -2)/np.exp(-x))/N
print(f'Integral pelo método de Importância = {integral:.4f}')

#plotar o grafico
x = np.linspace(0, 1000, N)
curve_y = x/(x**2 + 1)**2
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.xlim(0, 7); plt.xlabel('x'), plt.ylabel('y')
plt.text(4, 0.15, f'Integral = {integral:.4f}', fontsize=12, color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.show()

```
