```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100000
```

```python
def calcular_integral(u):
    x = 1 - np.sqrt(1 - u)
    return np.sum(pow(1-x**2, 3/2)/(2*(1-x)))/N
```

```python
# Calculando a integral por Monte Carlo
print("Valor da integral por Monte Carlo: {:.2f}".format(calcular_integral(np.random.rand(N))))
```

```python
# Calculando a integral por Importancia
print("Valor da integral por Importancia: {:.2f}".format(calcular_integral(np.random.uniform(0, 1, N))))
```

```python
x = np.linspace(0, 1000, N)
curve_y = pow(1 - x**2, 3/2)

plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.xlim(0, 1); plt.xlabel('x'), plt.ylabel('y')
plt.text(0.7, 0.9, f'Integral = {calcular_integral(np.random.uniform(0, 1, N)):.4f}', fontsize=12, color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
